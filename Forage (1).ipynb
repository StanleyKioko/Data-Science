```python
# importing libraries

import pandas as pd
```

```python
# laoding datasets
content_df = pd.read_csv('Content.csv')
reaction_df = pd.read_csv('Reactions.csv')
reactionType_df = pd.read_csv('ReactionTypes.csv')
```

## Content

```python
content_df.head()
```

```python
content_df.head()

print(content_df.columns)

# unique

print(content_df['Category'].unique())
print(content_df['ContentType'].unique())

# changing the casing of all values in the dataset

content_df[ 'Category'] = content_df[ 'Category'].str.lower()

print(content_df['Category'].unique())
print(content_df['ContentType'].unique())
```

## Reactions

```python
reaction_df.head()
```

```python
# convert datetime to datetime

print(reaction_df.dtypes)
reaction_df['Datetime'] = pd.to_datetime(reaction_df['Datetime'])
print(reaction_df.dtypes)



# extract year, hour, day of the week and month

# hour
reaction_df['Hour'] = pd.DatetimeIndex(reaction_df['Datetime']).hour

# Extracting day of week from date column

reaction_df['DayOfWeek'] = pd.DatetimeIndex(reaction_df['Datetime']).day_name()

# Extracting month from date column

reaction_df['Month'] = pd.DatetimeIndex(reaction_df['Datetime']).month_name()

# extracting the year from date column

reaction_df['Year'] = pd.DatetimeIndex(reaction_df['Datetime']).year


print(reaction_df.columns)
```

```python
# unique values
print(reaction_df['ReactionType'].unique())


# unique values
# print(reaction_df['DayOfWeek'].unique())
```

## reactiontype

```python
# columns

print(reactionType_df.columns)

print(reactionType_df.dtypes)

print(reactionType_df['ReactionType'].unique())
print(reactionType_df['Sentiment'].unique())
print(reactionType_df['Score'].unique())
```

## merging columns

```python
# sample -reaction df

reaction_df.sample(10)
```

```python
# Sample content df

content_df.sample(10)
```

```python
# sample - reaction type df

reactionType_df.sample(10)
```

```python
# pandas merging reaction df and content df on content Id

df = pd.merge(content_df, reaction_df, on='Content ID')

df.head()

print(len(reaction_df))
print(len(content_df))
print(len(df))

df.sample(10)
```

```python
# merging df and reaction type of reaction type

df1 = pd.merge(df, reactionType_df, on='ReactionType')

df1.head()

print(len(reaction_df))
print(len(content_df))
print(len(reactionType_df))
print(len(df1))
print(len(df))

df1.sample(10)
```

### saving cleaned datasheet

```python
# saving the dataframe
df1.to_csv('Solution1.csv')
```

## solution

```python
# top five categories

df_sum = df1.groupby('Category', sort=True)['Score'].sum().reset_index()
top5 = df_sum.sort_values(by = [ 'Score'], ascending=[False])

top5
```

## Group by


```python
reactions = (df1.groupby(['Category', 'ReactionType'], sort = False)['ReactionType'].agg('count').reset_index(name='Count'))
reactions
```

```python
df1.columns
```

```python
posts_per_month = (df1.groupby(['Content ID', 'Month'], sort = False)['Content ID'].agg('count').reset_index(name='Count'))
posts_per_month
```

```python
posts_per_day = (df1.groupby(['Content ID', 'DayOfWeek'], sort = False)['Content ID'].agg('count').reset_index(name='Count'))
posts_per_day
```

```python
# saving the dataframe
posts_per_day.to_csv('posts_per_day1.csv')
```

```python
# saving the dataframe
posts_per_month.to_csv('posts_per_month.csv')
```

```python
posts_per_hour = (df1.groupby(['Content ID', 'Hour'], sort = False)['Content ID'].agg('count').reset_index(name='Count'))
posts_per_hour
```

```python
# saving the dataframe
posts_per_hour.to_csv('posts_per_hour.csv')
```
